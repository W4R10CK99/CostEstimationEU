```python
import pandas as pd
```

```python
def add_years_after_installation(df):
    # Convert date columns to datetime objects with only the date part
    df['InstallDate'] = pd.to_datetime(df['InstallDate']).dt.date
    df['CreationDate'] = pd.to_datetime(df['CreationDate']).dt.date

    # Calculate the number of years after installation
    df['YearsAfterInstallation'] = (df['CreationDate'] - df['InstallDate']).dt.days // 365

    return df
```

```python
# Example usage:
# Load your dataset
df = pd.read_excel('dataset/DataFull.xlsx')
df.head()
```

```python
# Call the function to add the new column
updated_df = add_years_after_installation(df)
```
