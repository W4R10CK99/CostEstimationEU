```python
df_grouped = df.groupby(['AssetTypeID', df['CreationDate'].dt.year.rename('Year'), df['CreationDate'].dt.month.rename('Month')])['Spend_ExGST'].sum().reset_index()
```

```python
df_pivot = df_grouped.pivot_table(index=['Year', 'Month'], columns='AssetTypeID', values='Spend_ExGST', aggfunc='sum')

# Plot the total cost for AssetTypeID 15 on a monthly basis for each year
plt.figure(figsize=(12, 6))
df_monthly_15 = df_pivot[15]  # Extract data for AssetTypeID 15 from the pivot table

# Convert MultiIndex to a string representation
df_monthly_15.index = df_monthly_15.index.map(lambda x: f'{x[0]}-{x[1]}')

plt.bar(df_monthly_15.index, df_monthly_15)
plt.title('Monthly Total Cost for AssetTypeID 15 for Each Year')
plt.xlabel('Year-Month')
plt.ylabel('Total Cost')
plt.show()
```

```python
df_avg_spend = pd.read_excel('average_spend_by_asset_type.xlsx')
```

```python
train = df_avg_spend.loc[df_avg_spend['Year']<2023]
test = df_avg_spend.loc[df_avg_spend['Year']>=2023]
```

```python
train
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python
X_train = train[['AssetTypeID', 'Year', 'Month']]
y_train = train['Spend_ExGST']
X_test = test[['AssetTypeID', 'Year', 'Month']]
y_test = test['Spend_ExGST']
```

```python
reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                       n_estimators=1000,
                       early_stopping_rounds=50,
                       objective='reg:linear',
                       max_depth=3,
                       learning_rate=0.01)
```

```python
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)
```

```python
fi = pd.DataFrame(data=reg.feature_importances_,
             index=reg.feature_names_in_,
             columns=['importance'])
fi.sort_values('importance').plot(kind='barh', title='Feature Importance')
plt.show()
```

```python
test['prediction'] = reg.predict(X_test)
df = df_avg_spend.merge(test[['prediction']], how='left', left_index=True, right_index=True)
ax = df[['Spend_ExGST']].plot(figsize=(15, 5))
df['prediction'].plot(ax=ax, style='.')
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Dat and Prediction')
plt.show()
```

```python
# Filter the data for AssetTypeID == 15
train_15 = train[train['AssetTypeID'] == 15]
test_15 = test[test['AssetTypeID'] == 15]

# Features (X) and Target (y) for training set
X_train = train_15[['Year', 'Month']]
y_train = train_15['Spend_ExGST']

# Features (X) and Target (y) for testing set
X_test = test_15[['Year', 'Month']]
y_test = test_15['Spend_ExGST']
```

```python
reg2 = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                       n_estimators=1000,
                       early_stopping_rounds=50,
                       objective='reg:linear',
                       max_depth=3,
                       learning_rate=0.01)
```

```python
reg2.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)
```

```python
fi2 = pd.DataFrame(data=reg2.feature_importances_,
             index=reg2.feature_names_in_,
             columns=['importance'])
fi2.sort_values('importance').plot(kind='barh', title='Feature Importance')
plt.show()
```

```python
df = df_avg_spend[df_avg_spend['AssetTypeID']==15]
df = df[['Year','Month','Spend_ExGST']]
df.head()
```

```python
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))
df = df.sort_values('Date').reset_index(drop=True)

# Drop the 'Year' and 'Month' columns if needed
df = df.iloc[1:].reset_index(drop=True)
# Now 'Date' is the datetime column, and the data is sorted by date
print(df)
```

```python
df = df.set_index('Date')
df.index = pd.to_datetime(df.index)
```

```python
df['Spend_ExGST'].plot(style='.', figsize=(15, 5), title='Testing')
plt.xlabel('Date')
plt.ylabel('Spend_ExGST')
plt.show()
```

```python
train = df[df['Year']<2023]
test = df[df['Year']>=2023]
```

```python
# Features (X) and Target (y) for training set
X_train = train[['Year', 'Month']]
y_train = train['Spend_ExGST']

# Features (X) and Target (y) for testing set
X_test = test[['Year', 'Month']]
y_test = test['Spend_ExGST']
```

```python
reg3 = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                       n_estimators=1000,
                       early_stopping_rounds=50,
                       objective='reg:linear',
                       max_depth=3,
                       learning_rate=0.01)
```

```python
reg3.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)
```

```python
test['prediction'] = reg3.predict(X_test)
df = df.merge(test[['prediction']], how='left', left_index=True, right_index=True)
ax = df[['Spend_ExGST']].plot(figsize=(15, 5))
df['prediction'].plot(ax=ax, style='.')
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Dat and Prediction')
plt.show()
```
