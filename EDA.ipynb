```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel('Dataset/dataset.xlsx')
```

```python
dataset.head()
```

```python
#EDA
```

```python
ProblemType = dataset[['ProblemType']]
desc = ProblemType.value_counts()
plt.figure(figsize=(4, 4))
desc.plot.pie(autopct='%1.1f%%', startangle=90)

# Adding title
plt.title('Distribution of Problem Types')

# Display the plot
plt.show()
```

```python
CallType = dataset[['CallType']]
desc1 = CallType.value_counts()
plt.figure(figsize=(4, 4))
desc1.plot.pie(autopct='%1.1f%%', startangle=90)

# Adding title
plt.title('Distribution of Call Types')

# Display the plot
plt.show()
```
